3. Explain the following terms.

The difference in np.array, np.asarray and np.asanyarray
The difference between Deep copy and shallow copy

Ans :

np.array :- it is used to create 1d array.

np.asarray :- it is used to convert input into array.

np.asanyarray :- it convert any input to an ndarray.

Deep copy & shallow Copy

shallow copy :-

consider if we create list of element assigned variable v1. now we assigned v2 =v1 it has copied v1 list in v2. Now if we change index zero element of v2 it is reflecting in v1 also beacuse thise both list are pointing same memory location it is called as shallow copy.

Deep Copy :-

consider if we create list of element assigned variable v1. now we assigned v2=v1.copy() it has copied v1 list in v2. Now if we change index zero element of v2 it is not reflecting in v1 also beacuse thise both list are not pointing same memory location it is called as deep copy.

Example:

Shallow Copy

```python
lst1 =[1,2,3,4,5]

print(lst1)
```

```python
lst2 =lst1
print(lst2)
```

```python
lst2[0] =100

print(lst1)

print(lst2)
```

Deep Copy

```python
lst1 =[1,2,3,4,5]

print(lst1)
```

```python
lst3 =lst1.copy()

print(lst3)
```

```python
lst3[0] = 10

print(lst1)

print(lst3)
```

16 .Set the title of the plot as 'Histogram with PDF Overlay'.

```python
from scipy.stats import norm
#### Create 1000 sample of normal diastrubution data
data =np.random.normal(0,1,1000)

#### Plot histogram of bin=30
plt.hist(data,bins=30 ,label="Hist")

#### Overlay a line plot representing the normal distribution's probability density function (PDF)
xmin ,xmax =plt.xlim()
x=np.linspace(xmin ,xmax ,100)
p =norm.pdf(x ,np.mean(data),np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Step 4: Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

### set title
plt.title("Histogram with PDF Overlay")

plt.legend()
plt.show()
```

17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Generate two random arrays
np.random.seed(42)  # For reproducibility
x = np.random.uniform(-10, 10, 100)
y = np.random.uniform(-10, 10, 100)

#Determine the quadrant of each point
def determine_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Q1'
    elif x < 0 and y > 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    else:
        return 'Q4'

quadrants = [determine_quadrant(x[i], y[i]) for i in range(len(x))]

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y, 'Quadrant': quadrants})

#Create a scatter plot with Seaborn
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='bright', style='Quadrant', s=100)

# Add a legend, label the axes, and set the title
scatter_plot.set_title('Quadrant-wise Scatter Plot')
scatter_plot.set_xlabel('X-axis')
scatter_plot.set_ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Show the plot
plt.legend(title='Quadrant')
plt.show()
```

19. Using Bokeh, Generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import curdoc
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral6

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a Bokeh ColumnDataSource
source = ColumnDataSource(data)

#  Create a Bokeh figure
output_notebook()  # Use this if you're running in a Jupyter notebook
p = figure(x_range=categories, title="Random Categorical Bar Chart", toolbar_location=None, tools="")

# Plot the bar chart with colored bars
p.vbar(x='categories', top='values', width=0.9, source=source,
       legend_field="categories", line_color='white',
       fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

#  Add hover tooltips to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Set the title and adjust other properties
p.title.text = "Random Categorical Bar Chart"
p.legend.title = "Categories"
p.legend.orientation = "horizontal"
p.legend.location = "top_center"


show(p)
```
